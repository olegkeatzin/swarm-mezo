# Swarm-MeZO — K=10 consensus on RoBERTa+SST-2 (Colab)

Запускает `scripts/run_k10.py` на GPU — новая точка на оси `local_steps`.

**Идея.** Day 4/5 (репутация, trimmed) шли при `K=100` локальных шагов между
раундами консенсуса. При `K=100` агенты дрейфуют далеко на шуме MeZO, и
вопрос был — стабилен ли вообще probe-лосс как сигнал «кто плохой».
`K=10` синхронизирует рой в 10× чаще: агенты держатся теснее, но всё ещё
успевают разойтись достаточно, чтобы probe-лоссы `L_i` реально различались —
то есть у репутации/трима есть что ранжировать (в отличие от `K=1`, где
расхождение за один MeZO-шаг это чистый seed-шум).

Читать против `outputs/day5_trimmed.json` (`K=100`): тот же IID-split, та же
сетка, тот же `lr=1e-6` — меняется только `local_steps` (100 → 10).

**Прогон (IID):** N=8 агентов, `K=10`, 5000 шагов; сетка
`β ∈ {0, 0.1, 0.5, 1, 10} × trim_k ∈ {0, 2}` = **10 прогонов**
(`trim_k=0` — без трима: K=10 FedAvg при β=0, K=10 репутация при β>0).

**Время:** probe-eval каждые 10 шагов (500 раундов). ≈1.5–2ч на конфиг
(L4/A100), ≈3ч на T4. 10 конфигов = 15–30ч — несколько сессий (идемпотентно).

**Перед запуском:** (1) запушить `run_k10.py` в GitHub; (2) Runtime → GPU.

## 5. Быстрый превью результатов

Финальная val accuracy по сетке β × trim_k. Полный анализ — в отдельном
notebook позже.

```python
import json
import matplotlib.pyplot as plt

d = json.loads(open('outputs/day6_k10.json').read())
runs = d['runs']
betas = sorted({h['beta'] for h in runs.values()})
trims = sorted({h['trim_k'] for h in runs.values()})
print(f"{len(runs)} configs  |  β={betas}  trim_k={trims}  |  K={d['config']['local_steps']}")
print()
hdr = 'β \\ trim_k'
print(f'{hdr:>12} | ' + ' | '.join(f'k={k:>6}' for k in trims))
for b in betas:
    cells = []
    for k in trims:
        h = runs.get(f'beta{b}_trim{k}')
        cells.append(f'{h["eval_acc"][-1]:.4f}' if h else '   —  ')
    print(f'{b:>12} | ' + ' | '.join(f'{c:>8}' for c in cells))
```

```python
# Кривые val accuracy — одна линия на (β, trim_k)
fig, axes = plt.subplots(1, len(trims), figsize=(6 * len(trims), 5),
                         dpi=110, sharey=True, squeeze=False)
cmap = plt.get_cmap('viridis')
for ax, k in zip(axes[0], trims):
    sub = {b: runs.get(f'beta{b}_trim{k}') for b in betas}
    sub = {b: h for b, h in sub.items() if h}
    for i, (b, h) in enumerate(sub.items()):
        color = cmap(i / max(len(sub) - 1, 1))
        ax.plot(h['eval_step'], h['eval_acc'], marker='o', markersize=3,
                color=color, label=f'β = {b}')
    ax.set_title(f'trim_k = {k}' + (' (no trim)' if k == 0 else f' (drop worst {k}/8)'))
    ax.set_xlabel('per-agent step'); ax.grid(True, alpha=0.3); ax.legend()
axes[0][0].set_ylabel('SST-2 val accuracy')
fig.suptitle('K=10 consensus: federated MeZO on RoBERTa+SST-2 (IID)')
plt.tight_layout(); plt.show()
```
